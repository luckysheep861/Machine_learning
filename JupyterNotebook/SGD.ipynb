# 使用我们的SGD训练波士顿房价

```python
from sklearn import datasets

boston = datasets.load_boston()
X = boston.data
y = boston.target

X = X[y < 50.]
y = y[y < 50.]
```

```python
from playML.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, seed = 666)
```

```python
from playML.preprocessing import StandardScaler

stdsc = StandardScaler()
stdsc.fit(X_train)

X_train_std = stdsc.transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
from playML.LinearRegression import LinearRegression

lr_reg_sgd = LinearRegression()
lr_reg_sgd.fit_sgd(X_train_std, y_train, n_iters = 1)
lr_reg_sgd.score(X_test_std, y_test)
```

## scikit-learn 中的 SGD

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd_reg = SGDRegressor(n_iter = 100)
sgd_reg.fit(X_train_std, y_train)
sgd_reg.score(X_test_std, y_test)
```
