```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.,2.,3.,4.,5.])
y = np.array([1.,3.,2.,3.,5.])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from playML.SimpleLinearRegression import SimpleLinearRegression1
```

```python
my_slr = SimpleLinearRegression1()
```

```python
my_slr.fit(x,y)
```

```python
x_predict = 6
my_slr.predict(np.array([x_predict]))
```

```python
print(my_slr.a_)
print(my_slr.b_)
```

```python
y_hat = my_slr.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_hat,color='r')
plt.show()
```
